```python
import pandas as pd
import numpy as np

df = pd.read_csv("Datasets/lec06_weather_data.csv", parse_dates = ["day"])
df
```

```python
new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : -99999,
    'event' : '0'
}, np.NaN)

new_df
```

```python
new_df = df.replace({
    -99999 : np.NaN,
    '0': 'Sunny'
})
print(new_df)
```

```python
new_df = df.replace('[A-Za-z]','', regex=True)
new_df
```
